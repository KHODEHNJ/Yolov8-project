```
import cv2
import torch
import matplotlib.pyplot as plt
from ultralytics import YOLO

def detect_faces(image_path):

    model = YOLO('yolov8n.pt')


    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


    results = model(img_rgb)


    for result in results:
        for box in result.boxes.xyxy:
            x1, y1, x2, y2 = map(int, box[:4])
            cv2.rectangle(img_rgb, (x1, y1), (x2, y2), (255, 0, 0), 2)


    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()

detect_faces('/content/DALL·E 2025-03-27 00.19.38 - A realistic portrait of a young man with short brown hair and a confident expression. He has a strong jawline, light beard, and wears a stylish leathe.webp')
```

```
detect_faces("/content/actor.webp")
detect_faces("/content/Bam.webp")
detect_faces("/content/ahmad.jpg")
```
